```python
import autorootcwd
import json
import os
import pandas as pd
```

```python
DIR = './results/generative_features'
expts = os.listdir(DIR)
rows = []
for e in expts:
    with open(f"{DIR}/{e}/metadata.json") as f:
        args = json.load(f)

    args = args['args']
    reduction = args['gen_reduction']
    model_name = args['gen_model_name']

    if not os.path.exists(f"{DIR}/{e}/val_metrics.json"):
        print(f"Skipping {e} because val_metrics.json does not exist")
        continue
    
    with open(f"{DIR}/{e}/val_metrics.json") as f:
        data = json.load(f)

        rows.append({
            'reduction': reduction,
            'model_name': model_name,
            'accuracy': data['mean']['acc'],
            'root': data['mean']['root'],
            'majmin': data['mean']['majmin'],
            'third': data['mean']['third'],
            'seventh': data['mean']['seventh'],
            'mirex': data['mean']['mirex'],
            'mean_acc_class': data['class_wise']['acc']['mean'],
            'median_acc_class': data['class_wise']['acc']['median'],
            'avg_transitions': data['avg_transitions_per_song']
        })
 

df = pd.DataFrame(rows)
```

```python
df.sort_values(by=['accuracy'], inplace=True, ascending=False)
df
```

```python
DIR = './results/gen_down_dim'
expts = os.listdir(DIR)
rows = []
for e in expts:
    with open(f"{DIR}/{e}/metadata.json") as f:
        args = json.load(f)

    args = args['args']
    reduction = args['gen_reduction']
    model_name = args['gen_model_name']
    dim = args['gen_down_dimension']

    if not os.path.exists(f"{DIR}/{e}/val_metrics.json"):
        print(f"Skipping {e} because val_metrics.json does not exist")
        continue
    
    with open(f"{DIR}/{e}/val_metrics.json") as f:
        data = json.load(f)

        rows.append({
            'reduction': reduction,
            'model_name': model_name,
            'dim': dim,
            'accuracy': data['mean']['acc'],
            'root': data['mean']['root'],
            'majmin': data['mean']['majmin'],
            'third': data['mean']['third'],
            'seventh': data['mean']['seventh'],
            'mirex': data['mean']['mirex'],
            'mean_acc_class': data['class_wise']['acc']['mean'],
            'median_acc_class': data['class_wise']['acc']['median'],
            'avg_transitions': data['avg_transitions_per_song']
        })
 

df = pd.DataFrame(rows)
```
